```python
def split_atoms(s, multi_letter_atoms):
    atoms = []
    i = 0
    while i < len(s):
        if i + 1 < len(s) and s[i:i+2] in multi_letter_atoms:
            atoms.append(s[i:i+2])
            i += 2
        else:
            atoms.append(s[i])
            i += 1
    return atoms

def canonical_key(selected_atoms, multi_letter_atoms):
    atoms_fwd = split_atoms(selected_atoms, multi_letter_atoms)
    atoms_rev = list(reversed(atoms_fwd))
    key_fwd = ''.join(atoms_fwd)
    key_rev = ''.join(atoms_rev)
    return min(key_fwd, key_rev)
```

```python
multi_letter_atoms = ['Cl', 'Br', 'Sc', 'Ti', 'Cr', 'Mn', 'Fe', 'Co', 'Ni', 'Cu', 'Zn', 'Zr', 'Nb', 'Mo', 'Tc', 'Ru', 'Rh', 'Pd', 'Ag', 'Cd',
'La', 'Hf', 'Ta', 'Re', 'Os', 'Ir', 'Pt', 'Au', 'Hg']


def canonicalize_dict_keys(d, multi_letter_atoms):
    """
    Canonicalize dictionary keys while merging values for duplicate canonical keys
    """
    canonical_dict = {}
    
    for original_key, values in d.items():
        canonical = canonical_key(original_key, multi_letter_atoms)
        
        if canonical in canonical_dict:
            # Merge: extend the existing list with new values
            canonical_dict[canonical].extend(values)
        else:
            # First occurrence: create new list
            canonical_dict[canonical] = list(values)
    
    # Convert back to numpy arrays
    for key in canonical_dict:
        canonical_dict[key] = np.array(canonical_dict[key])
    
    return canonical_dict


```

```python

cleaned_OMol25_bonds = canonicalize_dict_keys(cleaned_OMol25_bonds, multi_letter_atoms)
cleaned_CCDC_bonds = canonicalize_dict_keys(cleaned_CCDC_bonds, multi_letter_atoms)
cleaned_CCDC_angles = canonicalize_dict_keys(cleaned_CCDC_angles, multi_letter_atoms)
cleaned_CCDC_torsions = canonicalize_dict_keys(cleaned_CCDC_torsions, multi_letter_atoms)
cleaned_HCSD_bonds = canonicalize_dict_keys(cleaned_HCSD_bonds, multi_letter_atoms)
cleaned_HCSD_angles = canonicalize_dict_keys(cleaned_HCSD_angles, multi_letter_atoms)
cleaned_HCSD_torsions = canonicalize_dict_keys(cleaned_HCSD_torsions, multi_letter_atoms)
cleaned_GOAT_bonds = canonicalize_dict_keys(cleaned_GOAT_bonds, multi_letter_atoms)
cleaned_GOAT_angles = canonicalize_dict_keys(cleaned_GOAT_angles, multi_letter_atoms)
cleaned_GOAT_torsions = canonicalize_dict_keys(cleaned_GOAT_torsions, multi_letter_atoms)
for geo in geometries:
    cleaned_GOAT_bonds_by_geo[geo] = canonicalize_dict_keys(cleaned_GOAT_bonds_by_geo[geo], multi_letter_atoms)
    cleaned_GOAT_angles_by_geo[geo] = canonicalize_dict_keys(cleaned_GOAT_angles_by_geo[geo], multi_letter_atoms)
    cleaned_GOAT_torsions_by_geo[geo] = canonicalize_dict_keys(cleaned_GOAT_torsions_by_geo[geo], multi_letter_atoms)
```

```python
import os
import json
import numpy as np

# Create directory if it doesn't exist
os.makedirs("Cleaned", exist_ok=True)

# Helper function to convert NumPy arrays to lists for JSON serialization
def convert_np_arrays(dictionary):
    converted_dict = {}
    for key, value in dictionary.items():
        if isinstance(value, np.ndarray):
            converted_dict[key] = value.tolist()
        else:
            converted_dict[key] = value
    return converted_dict


def save_all_files():
    try:
        # Save primary dictionaries
        with open("Cleaned/cleaned_OMol25_bonds.json", "w") as f:
            json.dump(convert_np_arrays(cleaned_OMol25_bonds), f)
        print(f"Saved cleaned_OMol25_bonds with {len(cleaned_OMol25_bonds)} entries")
        with open("Cleaned/cleaned_CCDC_bonds.json", "w") as f:
            json.dump(convert_np_arrays(cleaned_CCDC_bonds), f)
        print(f"Saved cleaned_CCDC_bonds with {len(cleaned_CCDC_bonds)} entries")
        with open("Cleaned/cleaned_CCDC_angles.json", "w") as f:
            json.dump(convert_np_arrays(cleaned_CCDC_angles), f)
        print(f"Saved cleaned_CCDC_angles with {len(cleaned_CCDC_angles)} entries")
        with open("Cleaned/cleaned_CCDC_torsions.json", "w") as f:
            json.dump(convert_np_arrays(cleaned_CCDC_torsions), f)
        print(f"Saved cleaned_CCDC_torsions with {len(cleaned_CCDC_torsions)} entries")
        with open("Cleaned/cleaned_HCSD_bonds.json", "w") as f:
            json.dump(convert_np_arrays(cleaned_HCSD_bonds), f)
        print(f"Saved cleaned_HCSD_bonds with {len(cleaned_HCSD_bonds)} entries")
        with open("Cleaned/cleaned_HCSD_angles.json", "w") as f:
            json.dump(convert_np_arrays(cleaned_HCSD_angles), f)
        print(f"Saved cleaned_HCSD_angles with {len(cleaned_HCSD_angles)} entries")  
        with open("Cleaned/cleaned_HCSD_torsions.json", "w") as f:
            json.dump(convert_np_arrays(cleaned_HCSD_torsions), f)
        print(f"Saved cleaned_HCSD_torsions with {len(cleaned_HCSD_torsions)} entries") 
        with open("Cleaned/cleaned_GOAT_bonds.json", "w") as f:
            json.dump(convert_np_arrays(cleaned_GOAT_bonds), f)
        print(f"Saved cleaned_GOAT_bonds with {len(cleaned_GOAT_bonds)} entries")
        with open("Cleaned/cleaned_GOAT_angles.json", "w") as f:
            json.dump(convert_np_arrays(cleaned_GOAT_angles), f)
        print(f"Saved cleaned_GOAT_angles with {len(cleaned_GOAT_angles)} entries")
        with open("Cleaned/cleaned_GOAT_torsions.json", "w") as f:
            json.dump(convert_np_arrays(cleaned_GOAT_torsions), f)
        print(f"Saved cleaned_GOAT_torsions with {len(cleaned_GOAT_torsions)} entries")
        # Create subdirectories for geometry-specific data
        os.makedirs("Cleaned/Geometries", exist_ok=True)
        
        # Save geometry-specific dictionaries
        for geo in geometries:
            geo_dir = os.path.join("Cleaned/Geometries", geo)
            os.makedirs(geo_dir, exist_ok=True)
            
            with open(f"{geo_dir}/bonds.json", "w") as f:
                json.dump(convert_np_arrays(cleaned_GOAT_bonds_by_geo[geo]), f)
                
            with open(f"{geo_dir}/angles.json", "w") as f:
                json.dump(convert_np_arrays(cleaned_GOAT_angles_by_geo[geo]), f)
                
            with open(f"{geo_dir}/torsions.json", "w") as f:
                json.dump(convert_np_arrays(cleaned_GOAT_torsions_by_geo[geo]), f)
    except Exception as e:
        print(f"Error: {str(e)}")

save_all_files()
```
